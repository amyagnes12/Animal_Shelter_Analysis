```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
import pandas as pd
```

```python
csv_path = "../Data/Austin_Animal_Center_Intakes.csv"
animal_center_intakes_df = pd.read_csv(csv_path)
animal_center_intakes_df
```

```python
animal_center_intakes_df.notnull().sum()
```

```python
# Drop unncessary columns
intakes_df = animal_center_intakes_df.drop(columns=["Name", "MonthYear", "Found Location", "Color"])
intakes_df
```

```python
# Get data for cat
intakes_df = intakes_df.loc[(intakes_df["Animal Type"] == "Cat")]
intakes_df
```

```python
intakes_df = intakes_df.dropna()
intakes_df.notnull().sum()
```

```python
# check if there are duplicates in animal ID
intakes_df['Animal ID'].is_unique
```

```python
# Drop duplicated animal ID
intakes_df = intakes_df.drop_duplicates(subset=['Animal ID'])
```

```python
intakes_df["DateTime"] = pd.to_datetime(intakes_df["DateTime"]).dt.date
intakes_df
```

```python
intakes_df.rename(columns = {'DateTime':'Intake Date'}, inplace = True)
```

```python
intakes_df["Age upon Intake"].unique()
```

```python
# Convert age into years
def convertAgeToFloat(inputAge):
# grab number & grab unit
    parsed = inputAge.split(" ")
    # ['5', 'days'] -> parsed[0] -> number, parsed[1] -> unit
    num = float(parsed[0])
    if (num < 0):
        return -1
    unit = parsed[1]
    divisor = 1
    if ('day' in unit):
        divisor = 365
    elif ('month' in unit):
        divisor = 12
    elif ('week' in unit):
        divisor = 52
    else:
        # assume the unit is years
        divisor = 1
    
    # if less than 1 year old, return 0
    if (num < divisor):
        return 0.0;
    
    # return age in years
    return num/divisor
```

```python
intakes_df["Age upon Intake"] = intakes_df["Age upon Intake"].apply(convertAgeToFloat)
intakes_df
```

```python
intakes_df["Age upon Intake"].unique()
```

```python
# Drop invalid age (-1)
intakes_df.drop(intakes_df.index[intakes_df["Age upon Intake"] == -1], inplace=True)
intakes_df
```

```python
# Rename age upon intake column
intakes_df.rename(columns = {'Age upon Intake':'Age upon Intake(years)'}, inplace = True)
```

```python
intakes_df['Breed'].unique()
```

```python
# Clean "Breed" column
# Remove breed name after "/" and change to "Mix"
intakes_df['Breed'] = intakes_df['Breed'].apply(lambda x: x.split("/")[0] + " Mix" if "/" in x else x)
intakes_df['Breed'].unique()
```

```python
intakes_df["Sex upon Intake"].unique()
```

```python
# Clean "Sex upon Intake" column
def convertSex(inputSex):
    if ('Male' in inputSex):
        return 'Male'
    elif ('Female' in inputSex):
        return 'Female'
    else: 
        return 'Unknown'    
```

```python
intakes_df["Sex upon Intake"] = intakes_df["Sex upon Intake"].apply(convertSex)
intakes_df
```

```python
# Drop unknown sex
intakes_df.drop(intakes_df.index[intakes_df["Sex upon Intake"] == "Unknown"], inplace=True)
intakes_df
```

```python
intakes_df["Intake Type"].unique()
```

```python
intakes_df["Intake Condition"].unique()
```

```python
# Clean "Intake Condition" column
def convertCondition(inputCondition):
    if ('Sick' in inputCondition) or ('Injured' in inputCondition) or ('Nursing' in inputCondition) or ('Medical' in inputCondition):
        return 'Sick'
    
    if ('Normal' in inputCondition):
        return 'Normal'
    
    if ('Aged' in inputCondition):
        return 'Aged'
    
    if ('Feral' in inputCondition):
        return 'Feral'
    
    if ('Pregnant' in inputCondition):
        return 'Pregnant'
    
    if ('Behavior' in inputCondition):
        return 'Behavior'
    
    return 'Other'        
```

```python
intakes_df["Intake Condition"] = intakes_df["Intake Condition"].apply(convertCondition)
intakes_df
```

```python
# Drop unknown condition
intakes_df.drop(intakes_df.index[intakes_df["Intake Condition"] == "Other"], inplace=True)
intakes_df
```

```python
intakes_df["Intake Condition"].unique()
```

```python
intakes_df.to_csv('../Data/clean_intakes.csv', index = False)  
```
