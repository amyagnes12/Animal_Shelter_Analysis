```python
import numpy as np
import pandas as pd
```

```python
csv_path = "Data/Austin_Animal_Center_Outcomes.csv"
animal_center_outcomes_df = pd.read_csv(csv_path)
animal_center_outcomes_df
```

```python
animal_center_outcomes_df.notnull().sum()
```

```python
# Drop unncessary columns
outcomes_df = animal_center_outcomes_df.drop(columns=["Name", "MonthYear", "Date of Birth", "Outcome Subtype", "Sex upon Outcome", "Breed", "Color"])
outcomes_df
```

```python
outcomes_df = outcomes_df.dropna()
outcomes_df.notnull().sum()
```

```python
# check if there are duplicates in animal ID
outcomes_df['Animal ID'].is_unique
```

```python
# Drop duplicated animal ID
outcomes_df = outcomes_df.drop_duplicates(subset=['Animal ID'])
```

```python
outcomes_df["DateTime"] = pd.to_datetime(outcomes_df["DateTime"]).dt.date
outcomes_df
```

```python
outcomes_df.rename(columns = {'DateTime':'Outcome Date'}, inplace = True)
```

```python
outcomes_df["Age upon Outcome"].unique()
```

```python
# Convert age into years
def convertAgeToFloat(inputAge):
# grab number & grab unit
    parsed = inputAge.split(" ")
    # ['5', 'days'] -> parsed[0] -> number, parsed[1] -> unit
    num = float(parsed[0])
    if (num < 0):
        return -1
    unit = parsed[1]
    divisor = 1
    if ('day' in unit):
        divisor = 365
    elif ('month' in unit):
        divisor = 12
    elif ('week' in unit):
        divisor = 52
    else:
    # assume the unit is years
        divisor = 1
    if (num < divisor):
        return 0.0;
    return num/divisor
```

```python
outcomes_df["Age upon Outcome"] = outcomes_df["Age upon Outcome"].apply(convertAgeToFloat)
outcomes_df
```

```python
outcomes_df["Age upon Outcome"].unique()
```

```python
# Drop invalid age (-1)
outcomes_df.drop(outcomes_df.index[outcomes_df["Age upon Outcome"] == -1], inplace=True)
outcomes_df
```

```python
# Rename age upon outcome column
outcomes_df.rename(columns = {'Age upon Outcome':'Age upon Outcomes(years)'}, inplace = True)
```

```python
outcomes_df["Animal Type"].unique()
```

```python
# Get data for dog and cat
outcomes_df = outcomes_df.loc[(outcomes_df["Animal Type"] == "Dog") | (outcomes_df["Animal Type"] == "Cat")]
outcomes_df
```

```python
# Drop animal type column
outcomes_df = outcomes_df.drop(columns=["Animal Type"])
```

```python
outcomes_df["Outcome Type"].unique()
```

```python
# Clean "Outcome Type" column
def convertOutcome(inputOutcome):
    if ('Adopt' in inputOutcome) or ('Return' in inputOutcome):
        return 'Adoption or RTO'
    elif ('Transfer' in inputOutcome) or ('Relocate' in inputOutcome):
        return 'Transfer'
    elif ('Died' in inputOutcome) or ('Disposal' in inputOutcome) or ('Missing' in inputOutcome):
        return 'Died or Missing'
    elif ('Euthanasia' in inputOutcome):
        return 'Euthanasia'
```

```python
outcomes_df["Outcome Type"] = outcomes_df["Outcome Type"].apply(convertOutcome)
outcomes_df
```

```python
outcomes_df["Outcome Type"].unique()
```

```python
outcomes_df.to_csv('Data/clean_outcomes.csv', index = False)  
```
