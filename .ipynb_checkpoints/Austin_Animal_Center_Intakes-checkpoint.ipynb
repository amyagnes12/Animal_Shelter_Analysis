```python
import numpy as np
import pandas as pd
```

```python
csv_path = "Data/Austin_Animal_Center_Intakes.csv"
animal_center_intakes_df = pd.read_csv(csv_path)
animal_center_intakes_df
```

```python
animal_center_intakes_df.notnull().sum()
```

```python
animal_center_df = animal_center_intakes_df.drop(columns=["Name", "MonthYear", "Found Location"])
animal_center_df
```

```python
animal_center_df = animal_center_df.dropna()
animal_center_df.notnull().sum()
```

```python
animal_center_df["DateTime"] = pd.to_datetime(animal_center_df["DateTime"]).dt.date
animal_center_df
```

```python
animal_center_df.rename(columns = {'DateTime':'Date'}, inplace = True)
```

```python
animal_center_df["Animal Type"].unique()
```

```python
# Convert age into years
def convertAgeToFloat(inputAge):
# grab number & grab unit
    parsed = inputAge.split(" ")
    # ['5', 'days'] -> parsed[0] -> number, parsed[1] -> unit
    num = float(parsed[0])
    if (num < 0):
        return -1
    unit = parsed[1]
    divisor = 1
    if ('day' in unit):
        divisor = 365
    elif ('month' in unit):
        divisor = 12
    elif ('week' in unit):
        divisor = 52
    else:
    # assume the unit is years
        divisor = 1
    if (num < divisor):
        return 0.0;
    return num/divisor
```

```python
for x in animal_center_df["Age upon Intake"]:
    print(convertAgeToFloat(x))
```

```python
animal_center_df["Age upon Intake"] = animal_center_df["Age upon Intake"].apply(convertAgeToFloat)
animal_center_df
```

```python
animal_center_df["Age upon Intake"].unique()
```

```python
# Drop invalid age (-1)
animal_center_df.drop(animal_center_df.index[animal_center_df["Age upon Intake"] == -1], inplace=True)
animal_center_df
```

```python
# Rename age upon intake column
animal_center_df.rename(columns = {'Age upon Intake':'Age upon Intake(years)'}, inplace = True)
```

```python
# Get data for dog and cat
animal_center_df = animal_center_df.loc[(animal_center_df["Animal Type"] == "Dog") | (animal_center_df["Animal Type"] == "Cat")]
animal_center_df
```

```python
animal_center_df["Sex upon Intake"].unique()
```

```python
# Clean "Sex upon Intake" column
def convertSex(inputSex):
    if ('Male' in inputSex):
        return 'Male'
    elif ('Female' in inputSex):
        return 'Female'
    else: 
        return 'Unknown'    
```

```python
animal_center_df["Sex upon Intake"] = animal_center_df["Sex upon Intake"].apply(convertSex)
animal_center_df
```

```python
# Drop unknown sex
animal_center_df.drop(animal_center_df.index[animal_center_df["Sex upon Intake"] == "Unknown"], inplace=True)
animal_center_df
```
